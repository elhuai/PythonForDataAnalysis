```python
import pandas as pd

df = pd.read_csv('各鄉鎮市區人口密度.csv')
df
```

```python
df1 = df.rename(columns={"statistic_yyy":"統計年",
                        "site_id":"區域別",
                        "people_total":"年底人口數",
                        "area":"土地面積",
                        "population_density":"人口密度"} )
df1
```

```python
df2 = df1.drop(index=0) #刪除第一列
df3 = df2.dropna(axis=0,how='any') #整理缺失值：每一列去遍歷有任何一個缺失值，就刪除整行
```

```python
df4 = df3.drop("統計年",axis=1) #刪除欄位
df4
```

```python
df4.info()
```

```python
# 清理資料
# 年底人口數 跟人口密度都改成整數
# 土地面積 改資料型態為浮點數
# apply(). 可以對每一個 series 中的每筆資料做函數執行
def strToInt(x:str)->int:
    try:
        return int(x)
    except:
        return 0
df4["年底人口數"] = df4["年底人口數"].apply(strToInt)
df4["人口密度"] = df4["人口密度"].apply(strToInt)
df4["土地面積"] = df4["土地面積"].astype("float")
df4
```

```python
df4.info() # 確定資料型態已改為數值，才可以做邏輯運算子
```

```python
def split_area(v):
    if len(v) >= 6:
        return v[:3]
    else:
        return v
    
def split_area1(v):
    if len(v) >= 6:
        return v[-3:]
    else:
        return v

df4["縣市"] = df4["區域別"].apply(split_area)
df4["區域"] = df4["區域別"].apply(split_area1)
df4
```

```python
df5 = df4.reindex(columns={"縣市","區域","年底人口數","土地面積","人口密度"})
```

```python
df5.query("年底人口數>=500000")
```

```python
m = df5["人口密度"].max()
df5.query("人口密度 == @m") #如果要帶入變數要加入@
```

```python
df6 = df5.groupby("縣市")[["年底人口數","土地面積","人口密度"]].agg([("平均","mean"),("加總","mean")])
df6
```

```python
df6.columns.names = ["資料","統計"]
```

```python
df6.stack(level = ['資料','統計'],future_stack=True.unstack(level=["縣市",'資料']))
```
