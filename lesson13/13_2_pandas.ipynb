```python
import numpy as np
import pandas
np.random.seed(10)
scores =  np.random.randint(50,101, size=(50,5))
df = pandas.DataFrame(scores,columns=["國","英","數","自","社"],index=[f"學生{i}" for i in range(1,51)])
# df是DataFrame的實體 (內含屬性跟方法)
# type(df)
```

```python
df.columns.name = "科目"
df.index.name = "姓名"
    
```

```python
#  新增總分欄位
df['總分'] = df[['國', '英', '數', '自', '社']].sum(axis=1)
df["平均"] = df["總分"]/5
df
```
