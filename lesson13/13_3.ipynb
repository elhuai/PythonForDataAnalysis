```python
import numpy as np
import pandas as pd

np.random.seed(10)
scores = np.random.randint(50,101,size=(50,5))
students = pd.DataFrame(scores,
             index=[f"學生{i}" for i in range(1,51)],
             columns=["國","英","數","自","社"]
)

# students[["國","英"]] #subscript 裡面放list
students_finish = students.copy() 
students_finish["總分"] = np.sum(students,axis=1) #axis=0 水平 axis=1垂直
students_finish
# students
```

```python
#橫列選取
students_finish.loc["學生1"] # 索引
students_finish.loc[["學生1","學生3"]] #利用[索引]
students_finish.loc["學生10":"學生30"] #利用slicing 快速切割
students_finish.loc["學生1":"學生10","國":"數"]  #第一筆資料列，第二列資料欄
students_finish.iloc[:7,:3]  #第一筆資料列，第二列資料欄
#slicing 都是一個view 會複製資料 不會更動到原始資料
students_finish.loc["學生1","國"]=100 #直接更改原始檔案位置
students_finish
```

```python
#利用dataFrame iloc loc 都會是複製的資料
#真的要多一份的時候要記得多一個.copy()

students10 = students_finish.iloc[:10].copy()
students10.loc["學生1","國"] = 100
students10
```

```python
#比較運算值
students_finish["英文優等"] = students_finish["英"]>=85


# 搜尋
# 英文大於85
students_finish[students_finish["英"]>=85]
# 總分大於500
students_finish[students_finish["總分"]>=400]
```
